### Store Part I results into DataFrame

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
# Store latitude and longitude in locations
locations = clean_df[["Lat", "Lng"]]

#Store Humidity in weight
weight = clean_df["Humidity"].astype(float)
weight
```

```python
fig = gmaps.figure(map_type = "HYBRID")

heat_layer = gmaps.heatmap_layer(locations, weights=weight, 
                                 dissipating=False, max_intensity=20,
                                 point_radius=1)

fig.add_layer(heat_layer)

fig
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = clean_df.copy()
hotel_df.head()
```

```python
# geocoordinates
lat = clean_df["Lat"][0]
lng = clean_df["Lng"][0]
target_coordinates = f"{lat}, {lng}"
#target_search = "Chinese"
target_radius = 5000
target_type = "hotel"
target_keyword = "hotel"


# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "keyword": target_keyword,
    "radius": target_radius,
    "type": target_type,
    "key": g_key
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
response = requests.get(base_url, params=params)
```

```python
print(response.url)
```

```python
# convert response to json
places_data = response.json()
places_data["results"][0]
```

```python
hotel_names = []

params = {
    "radius": 5000,
    "types": "hotel",
    "keyword": "hotel",
    "key": g_key
}

for index, row in hotel_df.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]

    params["location"] = f"{lat},{lng}"

    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # run a request
    response2 = requests.get(base_url, params=params)

    # convert to json
    hotels = response2.json()
    
    try:
        hotel_names.append(hotels["results"][0]["name"])
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        hotel_names.append("")
```

```python
#add the Hotel Name column to df
hotel_names
```

```python
hotel_df["Hotel Name"] = hotel_names
hotel_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{Cities}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
fig = gmaps.figure(map_type = "HYBRID")

coordinates = hotel_df[["Lat", "Lng"]]
marker_layer = gmaps.marker_layer(coordinates, info_box_content=hotel_info)
fig.add_layer(marker_layer)

# Display figure
fig
```
