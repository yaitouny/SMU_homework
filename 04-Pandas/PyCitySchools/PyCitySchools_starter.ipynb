```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_school = len(school_data_complete["school_name"].unique())
total_students = school_data_complete["Student ID"].count()
total_budget = school_data_complete.budget.unique().sum()

math_grades = school_data_complete["math_score"].sum()
reading_grades = school_data_complete["reading_score"].sum()
avg_math = math_grades / total_students
avg_reading = reading_grades / total_students

reading_passing = school_data_complete.loc[school_data_complete["reading_score"] >= 70]["reading_score"].count()
reading_percent = (reading_passing / total_students) * 100

math_passing = school_data_complete.loc[school_data_complete["math_score"] >= 70]["math_score"].count()
math_percent = (math_passing / total_students) * 100

overall_passing = school_data_complete.loc[(school_data_complete["reading_score"] >= 70) & 
                                           (school_data_complete["math_score"] >= 70)]
passing_count = overall_passing["Student ID"].count()
overall = (passing_count / total_students) * 100 



new_df = pd.DataFrame({"Total Schools": [total_school] , "Total Students": [total_students] ,
                       "Total Budget": [total_budget],
                      "Average Math Score":[avg_math], "Average reading score": [avg_reading]
                       , "% Passing Math":[math_percent] , "% Passing Reading": [reading_percent] ,
                      "% Overall Passing": [overall]})

new_df["Total Students"] = new_df["Total Students"].map("{:,}".format)
new_df["Total Budget"] = new_df["Total Budget"].map("{:,}".format)
new_df["Average Math Score"] = new_df["Average Math Score"].map("{0:.2f}".format)
new_df["Average reading score"] = new_df["Average reading score"].map("{0:.2f}".format)
new_df["% Passing Math"] = new_df["% Passing Math"].map("{0:.2f}%".format)
new_df["% Passing Reading"] = new_df["% Passing Reading"].map("{0:.2f}%".format)
new_df["% Overall Passing"] = new_df["% Overall Passing"].map("{0:.2f}%".format)


new_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
schools_df = school_data_complete.groupby("school_name")
#school_type = schools_df.type.unique()
school_type = [list[0] for list in schools_df.type.unique()]

student_count = schools_df["Student ID"].count()
school_budget = schools_df.budget.mean()

student_budget = school_budget / student_count
math_score = schools_df["math_score"].sum()
avg_math_score = math_score / student_count
reading_score = schools_df["reading_score"].sum()
avg_reading_score = reading_score / student_count

passing_read = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]

math_group = passing_read.groupby("school_name")
read_group = passing_math.groupby("school_name")

passed_m = (math_group["Student ID"].count() / student_count) * 100
passed_r = (read_group["Student ID"].count() / student_count) * 100 

passing_overall = school_data_complete.loc[(school_data_complete["reading_score"] >= 70) & 
                                           (school_data_complete["math_score"] >= 70)]
new_group = passing_overall.groupby("school_name")
passed_o = (new_group["Student ID"].count() / student_count) * 100 

school_summary = pd.DataFrame({"School Type": school_type, "Total Students": student_count, 
                               "Total School Budget": school_budget, "Per Student Budget": student_budget,
                              "Average Math Score": avg_math_score, "Average Reading Score":avg_reading_score, 
                              "% Passing Math":passed_m, "% Passing Reading": passed_r, "% Overall Passing": passed_o})

formatted_school_summary = school_summary.copy()

formatted_school_summary["Total Students"] = formatted_school_summary["Total Students"].map("{:,}".format)
formatted_school_summary["Total School Budget"] = formatted_school_summary["Total School Budget"].map("${:,}".format)
formatted_school_summary["Per Student Budget"] = formatted_school_summary["Per Student Budget"].map("${0:.2f}".format)
formatted_school_summary["Average Math Score"] = formatted_school_summary["Average Math Score"].map("{0:.2f}".format)
formatted_school_summary["Average Reading Score"] = formatted_school_summary["Average Reading Score"].map("{0:.2f}".format)
formatted_school_summary["% Passing Math"] = formatted_school_summary["% Passing Math"].map("{0:.2f}%".format)
formatted_school_summary["% Passing Reading"] = formatted_school_summary["% Passing Reading"].map("{0:.2f}%".format)
formatted_school_summary["% Overall Passing"] = formatted_school_summary["% Overall Passing"].map("{0:.2f}%".format)


formatted_school_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:

  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
bins = [0, 584, 629, 644, 675]
spending_labels = ["<$584", "$584-629", "$630-644", "$645-675"]

school_summary["Spending Ranges (Per Student)"] = pd.cut(school_summary["Per Student Budget"], bins, labels=spending_labels)
spending_group = school_summary.groupby("Spending Ranges (Per Student)")

avg_math_sp = spending_group["Average Math Score"].mean()
avg_read_sp = spending_group["Average Reading Score"].mean()
pass_math_sp = spending_group["% Passing Math"].mean()
pass_read_sp = spending_group["% Passing Reading"].mean()
overall_read_sp = spending_group["% Overall Passing"].mean()

spending_summary = pd.DataFrame({"Average Math Score": avg_math_sp, "Average Reading Score": avg_read_sp,
                               "% Passing Math": pass_math_sp, "% Passing Reading": pass_read_sp, 
                               "% Overall Passing": overall_read_sp})

                                 
formatted_spending_summary = spending_summary.copy()
formatted_spending_summary["Average Math Score"] = formatted_spending_summary["Average Math Score"].map("{0:.2f}".format)
formatted_spending_summary["Average Reading Score"] = formatted_spending_summary["Average Reading Score"].map("{0:.2f}".format)
formatted_spending_summary["% Passing Math"] = formatted_spending_summary["% Passing Math"].map("{0:.2f}%".format)
formatted_spending_summary["% Passing Reading"] = formatted_spending_summary["% Passing Reading"].map("{0:.2f}%".format)
formatted_spending_summary["% Overall Passing"] = formatted_spending_summary["% Overall Passing"].map("{0:.2f}%".format)

formatted_spending_summary
```
